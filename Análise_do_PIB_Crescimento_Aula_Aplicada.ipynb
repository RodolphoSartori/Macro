Resultado: O objeto datas_trimestrais é um DatetimeIndex contendo as datas que serão usadas para padronizar a coluna 'data' do DataFrame.

Resultado: A coluna 'data' agora contém as datas trimestrais no formato datetime.

```python
dados = pd.merge(df_mesclado1, df_selecionado6612, on='data', how='inner')
dados
```

```python
dados.dtypes
```

```python
dados['núm_índice_saz'] = pd.to_numeric(dados['núm_índice_saz'])
dados["núm_índice"] = pd.to_numeric(dados["núm_índice"])
```

Calculando as variadas taxas de crescimento

```python
#Trimestre contra treimestre imediantamente anterior (t contra t-1)
dados["var_trimestral"] = (dados["núm_índice_saz"] / dados["núm_índice_saz"].shift(periods = 1) - 1) * 100
#Trimestre contra mesmo Trimestre do ano Anterior (t contra t-4)
dados["var_interanual"] = (dados["núm_índice"] / dados["núm_índice"].shift(periods = 4) - 1) * 100
#Taxa acumulada em 4 trimestre contra o mesmo período do ano anterior (t contra t-12)
dados["var_anual"] = (dados["núm_índice"].rolling(4).sum()/dados["núm_índice"].shift(periods = 4).rolling(4).sum() - 1) * 100
dados
```

```python
#Criar coluna ano
dados['ano'] = pd.to_datetime(dados['data']).dt.year
#Soma acumulada dos números índices por ano
dados["núm_índice_acumulado"] = dados.groupby('ano')['núm_índice'].cumsum()
#Calcula a taxa de variação % do índice acumulado no ano
dados["var_ano_núm_índice_acumulado"] = (dados["núm_índice_acumulado"] / dados["núm_índice_acumulado"].shift(periods = 4) - 1)*100
dados
```

```python
dados.tail(15)
```

Visualizando os dados

```python
#Tratar os dados
dados = dados.melt(
    id_vars=['data'],
    value_vars=["var_trimestral", "var_interanual", "var_anual", "var_ano_núm_índice_acumulado"], #quais serão rotacionadas, o resto some
    value_name='valor_variável',
    var_name= 'variável',
).copy().replace(#recodificar valores das colunas
               to_replace = ["var_trimestral", "var_interanual", "var_anual", "var_ano_núm_índice_acumulado"],
               value = ["Variação Trimestral", "Variação Interanual", "Variação Anual", "var_ano_núm_índice_acumulado"]


)
dados
```

```python
#criar gráficos de linhas
fig = px.line(
    data_frame = dados,
    y = "valor_variável",
    x = "data",
    color = "variável",
    facet_col = "variável",
    facet_col_wrap = 2,
    title = "PIB: taxas de crescimento"
)

fig.update_yaxes(matches = None, title = "")
fig.update_xaxes(matches = None, title = "")
fig.for_each_annotation(lambda i: i.update(text = i.text.split("=")[-1]))
fig.update_layout(showlegend = False)
fig.add_annotation(
    showarrow = False,
    text = "Trimestre/ano",
    x = 0.5,
    xanchor = "center",
    xref= "paper",
    y = 0,
    yanchor = "bottom",
    yref = "paper",
    yshift = -30
)
fig.show()
```
